```python
import matplotlib.pyplot as plt
```

```python
with open('cora_timing.txt') as f:
    times = f.read()

times = [i.split('\t') for i in times[:-1].split('\n')]

x = [int(i[0]) for i in times[:-1]]
y = [int(i[1]) for i in times[:-1]]
```

```python

plt.figure(figsize=(10,6))
plt.rcParams['font.size'] = 16
plt.grid(True)
plt.xlabel("nnz_per_block")
plt.ylabel("Time (Microseconds)")
plt.title('Time taken to multiply Z = AX for Cora')
plt.plot(x, y)
```
